### Linear Regression Model

Now we're going to fit our data using [linear regression](https://en.wikipedia.org/wiki/Linear_regression). Linear regression fits two variables using a linear model. It tries to minimize the mean-squared error between the generated model and the actual data. By doing this, we try to see how we can predict the second variable given only the first.

```python
result = stats.linregress(combined_table[f"Total Value {var1_name}"], combined_table[f"Total Value {var2_name}"])
slope, intercept, r, p, se_slope, se_intercept = result.slope, result.intercept, result.rvalue, result.pvalue, result.stderr, result.intercept_stderr
plt.scatter(combined_table[f"Total Value {var1_name}"], combined_table[f"Total Value {var2_name}"], label='Data Points')
plt.plot(combined_table[f"Total Value {var1_name}"], intercept + slope*combined_table[f"Total Value {var1_name}"], 'r', label='fitted line')
plt.xlabel(f"{var1_name} Value")
plt.ylabel(f"{var2_name} Value")
plt.title(f"{var1_name} vs {var2_name} Fitted Plot")
plt.legend()
plt.show()
```

### Polynomial Regression Model

In this section, we are going to use a more powerful model than linear regression, namely using polynomials. Here, we have a tuneable hyperparameter $n$, which signifies the order of the polynomial. For instance, if $n = 2$, we are finding the best $a, b, c$ such that our model $$f(x) = a x^2 + b x + c$$ predicts $y$ as closely as possible based on our data.

This is actually the same as linear regression if we set $n$ correctly. What is $n$ in linear regression?

ANSWER: [YOUR ANSWER HERE]

First, set $n$.

```python
n = 2
```

Now, we run the regression itself.

```python
from scipy import odr
import numpy as np

x, y = combined_table[f"Total Value {var1_name}"], combined_table[f"Total Value {var2_name}"]
poly_model = odr.polynomial(n)
data = odr.Data(x, y)
odr_obj = odr.ODR(data, poly_model)
output = odr_obj.run()
poly = np.poly1d(output.beta[::-1])
x_sorted = sorted(x)
poly_y = poly(x_sorted)
plt.scatter(x, y, label="data")
plt.plot(x_sorted, poly_y, label="polynomial fit", color='r')
plt.legend()
plt.show()
```
