## Loading and Visualizing Your Data

### Loading the Data

To start, we need to load in our data. We have provided sample data for you, but we recommend you upload your own data to this collab environment and modify `var1_name`, `var2_name` and `var1_file`, `var2_file` to reflect your own data.

Your data should match the format of the sample data `sleep.csv` and `step.csv`. They should be csv files two columns: `Date` and `Total Value`.

```python
import pandas as pd

var1_name = "Sleep"
var2_name = "Step"
var1_file = "sample_data/Sleepanalysis_summary.csv"
var2_file = "sample_data/StepCount_summary.csv"

var1 = pd.read_csv(var1_file)
var2 = pd.read_csv(var2_file)
```

If all is well, this cell should have loaded both of your csv files. Let's see what the data looks like!

```python
print(f"A subset of {var1_name} data:")
var1.head()
```

```python
print(f"A subset of {var2_name} data:")
var2.head()
```

### Visualizing the Data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))  # Set the figure size

plt.plot(var1['Date'], var1['Total Value'], label='Minutes Asleep', color='blue')
plt.ylabel('Sleep (minutes)')
plt.xlabel('Date')
plt.title('Daily sleep minutes')
plt.show()
```

Now, we're going to try to visualize the data together. Since each of our data points has a date and a value, we're going to group together points by date.

```python
combined_table = pd.merge(var1, var2, on="Date", suffixes=[' ' + var1_name, ' ' + var2_name])
combined_table.head()
```

```python
plt.scatter(combined_table[f"Total Value {var1_name}"], combined_table[f"Total Value {var2_name}"])
plt.xlabel(f"{var1_name} Value")
plt.ylabel(f"{var2_name} Value")
plt.title(f"{var1_name} vs {var2_name} Scatter Plot")
plt.show()
```

### t-test

Next, we're going to try to do an analysis on our data using a t-test. Recall that a t-test is a statistical test used for evaluating whether the relationship between two groups is significant or not.
```python
from scipy import stats
stats.ttest_ind(combined_table[f"Total Value {var1_name}"], combined_table[f"Total Value {var2_name}"])
```
